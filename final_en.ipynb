````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

# importing Qiskit
from qiskit import IBMQ, BasicAer
from qiskit.providers.ibmq import least_busy
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister, execute,transpile

# import basic plot tools
from qiskit.tools.visualization import plot_histogram

def week3_ans_func(problem_set):
    ##### build your quantum circuit here
    ##### In addition, please make it a function that can solve the problem even with different inputs (problem_set). We do validation with different inputs. 
    
    address = QuantumRegister(4, name="address")
    board = QuantumRegister(16, name="board_main")
    counter = QuantumRegister(3, name="counter")
    oracle_bit = QuantumRegister(1, name="oracle")
    #oracle_bit2 = QuantumRegister(1, name="oracle_2")
    measure = ClassicalRegister(4,name="measure")
    
    qc = QuantumCircuit(address,board,counter,oracle_bit,measure)
    keys = ""
    dict_board = {"00":0,"01":1,"02":2,"03":3,"10":4,"11":5,"12":6,"13":7,"20":8,"21":9,"22":10,"23":11,"30":12,"31":13,"32":14,"33":15}
    
    qc.x(oracle_bit)
    qc.h(oracle_bit)
    qc.h(address)
    #qc.h(board)
    
    #Store board values as indexes
    Board_init = []
    for coordinate in problem_set:
        xy = []
        for coord in coordinate:
            xy.append((int(coord[0])*4)+int(coord[1]))
        Board_init.append(xy)
        
    #Mark index on binary board
    Board = [[0]*16 for j in range(16)]
    
    for val in range(16):
        for index in Board_init[val]:
            Board[val][index] = 1
            
    #Set horizontal,vertical vals for beams
    hor0 = set([12,13,14,15])
    hor1 = set([8,9,10,11])
    hor2 = set([4,5,6,7])
    hor3 = set([0,1,2,3])
    vert0 = set([0,4,8,12])
    vert1 = set([1,5,9,13])
    vert2 = set([2,6,10,14])
    vert3 = set([3,7,11,15])
    Beams = [hor0,hor1,hor2,hor3,vert0,vert1,vert2,vert3]
    
    def To_Binary(val,length):
        binary = []
        for j in range(length):
            binary.append(val%2)
            val = val//2
        binary.reverse()
        return binary
    
    
    #Store 3 shot beams
    BeamPewPew = []
    for val in range(16*16):
        valid_beam = To_Binary(val,8)
        if valid_beam.count(1) == 3:
            BeamPewPew.append(valid_beam)
    
    #Store values not broken by BeamPewPew
    Missed = []
    for PewPew in BeamPewPew:
        boop = set([val for val in range(16)])
        for j in range(8):
            if PewPew[j] == 1:
                boop = boop - Beams[j]
        Missed.append(list(boop))

    def qRam():
        #QRAM address -> 0
        key = ""
        qc.x(address[0:4])
        for problem in problem_set[0]:
            for value in problem:
                key += value
            qc.mct([address[0],address[1],address[2],address[3]],board[(dict_board[key])])
            key = ""
        
        qc.x(address[0:4])
        qc.barrier()
        #QRAM address -> 1
        qc.x(address[1:4])
        for problem in problem_set[1]:
            for value in problem:
                key += value
            qc.mct([address[0],address[1],address[2],address[3]],board[(dict_board[key])])
            key = ""
        qc.x(address[1:4])
        qc.barrier()    
        
      #QRAM address -> 2          
        qc.x(address[0])
        qc.x(address[2:4])
        for problem in problem_set[2]:
            for value in problem:
                key += value
            qc.mct([address[0],address[1],address[2],address[3]],board[(dict_board[key])])
            key = ""
        qc.x(address[0])
        qc.x(address[2:4])
        qc.barrier()
        
        #QRAM address -> 3         
        qc.x(address[2:4])
        for problem in problem_set[3]:
            for value in problem:
                key += value
            qc.mct([address[0],address[1],address[2],address[3]],board[(dict_board[key])])
            key = ""
        qc.x(address[2:4])
        qc.barrier() 
        
        #QRAM address -> 4        
        qc.x(address[0:2])
        qc.x(address[3])
        for problem in problem_set[4]:
            for value in problem:
                key += value
            qc.mct([address[0],address[1],address[2],address[3]],board[(dict_board[key])])
            key = ""
        qc.x(address[0:2])
        qc.x(address[3])
        qc.barrier()
        
        #QRAM address -> 5        
        qc.x(address[1:2])
        qc.x(address[3])
        for problem in problem_set[5]:
            for value in problem:
                key += value
            qc.mct([address[0],address[1],address[2],address[3]],board[(dict_board[key])])
            key = ""
        qc.x(address[1:2])
        qc.x(address[3])
        qc.barrier()
        
        #QRAM address -> 6        
        qc.x(address[0])
        qc.x(address[3])
        for problem in problem_set[6]:
            for value in problem:
                key += value
            qc.mct([address[0],address[1],address[2],address[3]],board[(dict_board[key])])
            key = ""
        qc.x(address[0])
        qc.x(address[3])
        qc.barrier()
        
        #QRAM address -> 7        
        qc.x(address[3])
        for problem in problem_set[7]:
            for value in problem:
                key += value
            qc.mct([address[0],address[1],address[2],address[3]],board[(dict_board[key])])
            key = ""
        qc.x(address[3])
        qc.barrier()
        
        #QRAM address -> 8        
        qc.x(address[0:3])
        for problem in problem_set[8]:
            for value in problem:
                key += value
            qc.mct([address[0],address[1],address[2],address[3]],board[(dict_board[key])])
            key = ""
        qc.x(address[0:3])
        qc.barrier()
        
        #QRAM address -> 9       
        qc.x(address[1:3])
        for problem in problem_set[9]:
            for value in problem:
                key += value
            qc.mct([address[0],address[1],address[2],address[3]],board[(dict_board[key])])
            key = ""
        qc.x(address[1:3])
        qc.barrier()
        
        #QRAM address -> 10        
        qc.x(address[0])
        qc.x(address[2])
        for problem in problem_set[10]:
            for value in problem:
                key += value
            qc.mct([address[0],address[1],address[2],address[3]],board[(dict_board[key])])
            key = ""
        qc.x(address[0])
        qc.x(address[2])
        qc.barrier()
        
        #QRAM address -> 11        
        qc.x(address[2])
        for problem in problem_set[11]:
            for value in problem:
                key += value
            qc.mct([address[0],address[1],address[2],address[3]],board[(dict_board[key])])
            key = ""
        qc.x(address[2])
        qc.barrier()
        
        #QRAM address -> 12        
        qc.x(address[0])
        qc.x(address[1])
        for problem in problem_set[12]:
            for value in problem:
                key += value
            qc.mct([address[0],address[1],address[2],address[3]],board[(dict_board[key])])
            key = ""
        qc.x(address[0])
        qc.x(address[1])
        qc.barrier()
        
        #QRAM address -> 13        
        qc.x(address[1])
        for problem in problem_set[13]:
            for value in problem:
                key += value
            qc.mct([address[0],address[1],address[2],address[3]],board[(dict_board[key])])
            key = ""
        qc.x(address[1])
        qc.barrier()
        
        #QRAM address -> 14        
        qc.x(address[0])
        for problem in problem_set[14]:
            for value in problem:
                key += value
            qc.mct([address[0],address[1],address[2],address[3]],board[(dict_board[key])])
            key = ""
        qc.x(address[0])
        qc.barrier()
        
        #QRAM address -> 15        
        for problem in problem_set[15]:
            for value in problem:
                key += value
            qc.mct([address[0],address[1],address[2],address[3]],board[(dict_board[key])])
            key = ""
        qc.barrier()
        
    def oracle(miss):
        #Flip value of missed
        for val in miss:
            qc.x(board[val])
        qc.barrier()
        #apply counter
        counter_func(miss)
        #flip again
        for val in miss:
            qc.x(board[val])
        qc.barrier()
        
        
    def rev_oracle(miss):
        for val in miss:
            qc.x(board[val])
        qc.barrier()
        #apply counter
        rev_counter(miss)
        #flip again
        for val in miss:
            qc.x(board[val])
        qc.barrier()
        
    def counter_func(miss):
        controller = []
        
        for val in miss:
            controller.append(board[val])
            
        controller.append(counter[0])
        qc.ccx(counter[0],counter[1],counter[2])
        qc.mct(controller,counter[1])
        qc.mct(controller[:-1],counter[0])
        qc.barrier()
            
    def rev_counter(miss):
        controller = []
        
        for val in miss:
            controller.append(board[val])
        
        controller.append(counter[0])
        qc.mct(controller[:-1],counter[0])
        qc.mct(controller,counter[1])
        qc.ccx(counter[0],counter[1],counter[2])
        qc.barrier()
    
    def diffusion():
        qc.h(address)
        qc.x(address)
        qc.h(address[0])
        qc.mct(address[1:], address[0], mode='noancilla')
        qc.h(address[0])
        qc.x(address)
        qc.h(address)
        
    #### Code for Grover's algorithm with iterations = 1 will be as follows. 
    #qRam
    qRam()
    qc.barrier()
    
    #Oracle
    for miss in Missed:
        oracle(miss)
    qc.barrier()
    
    #Counter    
    qc.x(counter)
    qc.mct(counter, oracle_bit)
    qc.x(counter)
    qc.barrier()
    
    #Rev_Counter
    for miss in Missed:
        rev_oracle(miss)
    qc.barrier()
    
    #qRam
    qRam()
    
    #Diffusion
    diffusion()
      
    qc.barrier()
    qc.h(oracle_bit)
    qc.x(oracle_bit)
    qc.barrier()
    
    qc.measure(address,measure)
    qc=qc.reverse_bits()
    
    from qiskit.transpiler import PassManager
    from qiskit.transpiler.passes import Unroller
    pass_ = Unroller(['u3', 'cx'])
    pm = PassManager(pass_)
    new_circ = pm.run(qc)
    qc = transpile(new_circ, basis_gates=['u3','cx'],optimization_level=3)
    
    return qc

#qc =week3_ans_func(problem_set)
#qc.draw()
````
